**Checking whether data is Stationary OR Not**

**ADFULLER TEST**

```python
from statsmodels.tsa.stattools import adfuller
```

Let, Ho - Data is not stationary ,
     H1 - Data is stationary

```python
def adf_test(data):
    if adfuller(data)[1]>0.05:
        print('accept ho, so data is not stationary')
    else:
        print('reject ho, so data is  stationary')
```

```python
adf_test(data['Close'])
```

**Train & Test data sets**

```python
train=new_data[:4300]
test=new_data[4300:]
```

```python
plt.plot(train,color='green')
plt.plot(test,color='red')
plt.show()
```

```python
def prepare_data(data,n_features):
    x,y = [],[]
    for i in range(len(data)):
        end_ix = i + n_features
        if end_ix > len(data) - 1:
            break
        seq_x,seq_y = data[i:end_ix],data[end_ix]
        x.append(seq_x)
        y.append(seq_y)
    x = np.array(x)
    y = np.array(y)

    # Flatten each sequence in x to create a 2D array
    x = x.reshape((x.shape[0],-1))

    return x,y
```

```python
x_train,y_train=prepare_data(train,70)
```

```python
x_test,y_test=prepare_data(test,70)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```
